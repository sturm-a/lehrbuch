# Theoretischer Hintergrund

`ggplot2` (R) und `ggplot` (Python) sind Pakete um eine grafische Darstellung von Plots zu erleichtern.
Wobei als Plot die grafische Darstellung von Zusammenhängen zwischen zwei oder mehrerer Variablen eines Datensatzes gemeint ist. 

Das Paket `ggplot` von Python basiert dabei auf dem Paket `ggplot2` von R, wobei beide Pakete sich dafür an die Leitlinien des Buches 'The Grammar of Graphics' von Wilkinson halten. Dieses Buch wurde geschrieben, um die Frage zu beantworten was eine statistische Grafik ist. Definiert wird eine Grafik in 'The Grammar of Graphics' als ein
Zusammenschluss von Daten, Attributen sowie geometrische Objekte. Attribute sind Merkmale, welche die Farbe, das Aussehen und die Form einer Grafik beschreiben und geometrische Objekte sind zum Beispiel Punkte, Linien oder Balken.

Da das Paket `ggplot` von Python auf dem von R basiert, ist der theoretische Hintergrund beides mal der selbe.

### Unterschiede zwischen `ggplot` und `ggplot2`

Die Unterschiede sind auf den ersten Blick eher minimal und auf die Schreibweise des Codes bezogen. Dies liegt vorallem daran, dass `ggplot` von Python auf dem Paket `ggplot2` von R basiert. Natürlich gibt es auch teilweise Funktionen, welche in R vorkommen, aber in Python nicht. Ein Beispiel hierfür ist die Kartenfunktion in R. Über die Layer `geom_polygon` und `coord_quickmap` kann in R eine vereinfachte Karte dargestellt werden. Jedoch muss man sich hierfür erstmal die benötigten Koordinaten laden.
Eine weitere Funktion welche nur in Python auftaucht, ist eine Spaßfunktion namens `geom_now_its_art`. Egal welcher Datensatz verwendet wird, am Ende wird immer ein Vogel als Bild eingezeichnet und die eigentlichen Daten überschrieben.

### Was ist alles möglich?

Hier sind ein paar Beispiele, was genau in R und Python dargestellt werden kann:

- Horizontale, vertikale oder diagonale Referenzlinien
- Balkendiagramme
- Heatmap
- Box/Whiskerplot
- Punkte
- Histogramme
- Scatterplots
- Einfache Texte
- u.a.

Es ist also zu erkennen, dass ggplot/ggplot2 eine sehr mächtige Funktion ist, welche einen Datensatz in sehr vielen, verschiedenen Arten darstellen kann. Dabei kann viel an dem Aussehen des Plots manuell geändert werden.

### Testdatensätze:

Zusammen mit dem Paket können jeweils auch Testdatensätze geladen werden, um sich in die jeweiligen Pakete einarbeiten zu können. Dabei sind viele Datensätze sowohl in Python als auch in R zu verwenden, z.B 
    
- Iris Datensatz: 
    Dieser Datensatz beinhaltet Daten von drei verschiedenen Arten der Iris-Pflanze sowie die Abmessung der Länge/Breite des Blüten- bzw. Kelchblattes.
    
- mpg Datensatz:
    Beinhaltet Informationen über die Kraftstoffeffizienz von beliebten Automarken
- diamonds Datensatz: Beinhaltet über 50.000 Daten von Diamanten unter anderem mit Angaben zum Preis und Karat.
- u.a.

### Schlüsselkomponenten:

Jede Grafik hat drei Schlüsselkomponenten, auf die der Plot aufgebaut wird:

- Ein Datensatz, der verwendet wird um die Grafik zu zeichnen.

- Ein Satz von `aesthetic mappings`: Dadurch wird angegeben, wie Variablen visuell angezeigt werden sollen. Diese Funktion wird in beiden Paketen mit `aes()` aufgerufen.

- Mindestens ein `layer`, um anzugeben wie der Datensatz gezeichnet werden soll.

Vereinfacht dargestellt, besteht ein Plot also jeweils aus drei Teilen:

*** Plot = (data + aestetic mapping) + layer ***

Dabei wird bei der Erstellung eines Plots zuerst über die Komponente `data` ein Datensatz zugewiesen und über die Komponente `aesthetic mappings` kann die gewünschte Darstellung von z.B. der Farbe und der Achsenbeschriftung gewählt werden. Nun können in weiteren Schritten ein oder mehrere `layer` hinzugefügt werden.

Beispiel eines einfachen Plots:

*R:*

```R
ggplot(mpg, aes(displ, hwy)) + geom_point()
```

*Python:*
```Python
ggplot(mpg, aes('displ', 'hwy')) + geom_point()
```

### Die wichtigsten Funktionen von aes():

Zuallererst wird in der aes()-Funktion angegeben, welche Variablen für die x/y-Achse verwendet werden.
Dies kann auf unterschiedliche Weisen angegeben werden und hier gibt es einen der Unterschiede in der Schreibweise zwischen Python und R:


*R:*
```R 
aes(displ, hwy)
aes(x=displ, y=hwy)
```

*Python:*
```Python
aes('displ', 'hwy')
aes(x='displ', y='hwy')
```



Weitere wichtige Optionen von `aes()` sind:

- **colour/color**: Die Ausgabe wird über eine dritte Variable farblich gruppiert und eine Legende erscheint am Rand.
- **shape**: Die Ausgabe wird über eine dritte Variable über geometrische Formen gruppiert und eine Legende erscheint am Rand. Jedoch kann es in der angegebenen Gruppe nicht mehr als 6 verschiedene Ausprägungen geben.
- **size**: Hierüber kann die Größe der Punkte/Linien angegeben werden, defaul ist größtenteils 0,5. Jedoch kann es auch als eine Gruppierungsmöglichkeit angesehen werden, dann werden für jede Gruppe die Punkte in unterschiedlich großen Punkte angezeigt.

    *R:*
    ```R 
    aes(displ, hwy, color=class)
    ```

    *Python:*
    ```Python
    aes('displ', 'hwy', color='class')
    ```

### Was ist ein Layer?

Ein Layer besteht aus fünf verschiedenen Komponenten.

1. `mapping`:
Dies ist wie bereits oben erwähnt ein Satz von `aesthetic mappings`. Als default werden hierbei die `aesthetic mappings` verwendet die bei der Erstellung des neuen Plots angegeben wurden. Es können jedoch auch für jeden `layer` eigene Eigenschaften definiert werden, bzw. die Angaben die am Anfang erstellt wurden überschrieben werden.

2. `data`:
Auch hier ist der default Datensatz derjenige, der bei der Erstellung des Plots angegeben wurde. Jedoch kann für jedes `layer` auch ein eigener Datensatz verwendet werden.

3. `geom`:
Ein geometrisches Element, welches angibt wie genau die Daten dargestellt werden sollen. Man kann unter anderem Linien- oder Balkendiagramme wählen, sowie noch verschiedene andere Optionen, welche in der praktischen Ausarbeitung erwähnt werden werden.

4. `stat`:
Eine statistische Transformation welche verwendet wird, um die Daten sinnvoll zusammenzufassen, z.B über einen Boxplot oder ein Histogramm.

5. `position`:
Eine Methode die benutzt wird, um überlappende Objekte anzupassen.

Wenn mehrere `layer` in einem Plot verwendet werden, müssen die Datensätze sowie die `aesthetic mappings` nicht die selben sein, sondern können innerhalb der verschiedenen `layer` individuell verändert werden.

###  Scales

Scales steuern die Abbildung der Daten in einen Plot. Über sie können Veränderungen im Aussehen eines Plots vorgenommen werden. Sollte in einem Code kein Scale explizit angegeben werden, läuft im Hintergrund trotzdem eine "leere" default Funktion.

Angewendet auf das Bespiel von oben für einen einfachen Plot, bedeutet dies:

*R:*
```R
ggplot(mpg, aes(displ, hwy)) + 
    geom_point() + 
    scale_x_continuous() +
    scale_y_continuous()
```

*Python:*
```Python
ggplot(mpg, aes('displ', 'hwy')) + \
    geom_point() + \
    scale_x_continuous() + \
    scale_y_continuous()
```

> Da diese zwei Scales häufig vorkommen, gibt es für sie auch eine Abkürzung, welche in Python und R jeweils die selbe ist:

> ```R
xlab()
ylab()
```

> Beziehungsweise noch vereinfachter ist die Schreibweise:

> ```R
labs(x= , y= )
```

### Legenden

Legenden sind ein wichtiger Bestandteil vieler Grafiken und können Angaben von verschiedenen `Layer` sowie auch unterschiedliche `aesthetic mappings` beinhalten. Sie können an unterschiedlichen Stellen des Plots dargestellt werden.

### Facetting

Hierbei kann ein großer Plot aufgeteilt werden in mehrere Plots. Dabei wird nach einer bestimmten Variable gruppiert, die angegeben werden kann. Wie genau dies Funktioniert wird später im praktischen Teil erklärt.

### Variablenzuweisung

Die Funktion *ggplot()* kann auch einer Variablen zugewiesen werden, sodass jeweilige Änderungen Schritt für Schritt hinzugefügt werden können. Wichtig ist, dass in der Erstinitialisierung der Teilcode *ggplot()* mit der Angabe des zu verwendeten Datensatzes vorkommt. Jedoch gibt es hierbei einen gravierenden Unterschied zwischen R und Python. Wobei es in R ausreicht, nur den Datensatz in der Funktion *ggplot()* anzugeben, muss in Python noch die leere *aes()* Funktion hinzugefügt werden. Layer und andere Optionen können dann im späteren Verlauf mit einem einfachen **+** hinzugefügt werden.

*R:*
```R
plot <- ggplot(mpg) 
    
plot + geom_point(aes(displ, hwy))
```

*Python:*
```Python
plot = ggplot(mpg, aes())

plot + geom_point(aes('displ', 'hwy'))
```

# Praktischer Hintergrund

Auch wenn im weiteren Verlauf jeweils Codes für beide Pakete angezeigt werden, wird jeweils nur der Code für `ggplot2` ausgeführt, da dieses Notebook R verwendet.
Es wird der Testdatensatz *diamonds* verwendet, wobei in der unten angefügten Tabelle zu erkennen ist, welche Variablen in diesem Datensatz vorkommen.

````r
head(diamonds)
````

Nun können wir mit einem Beispiel starten. Zu Beginn soll erst einmal dargestellt werden, ob es eine Beziehung zwischen dem Preis und Karat gibt, dies kann über ein Scatterplot dargestellt werden:

````r
library(ggplot2)
ggplot(diamonds, aes(x=carat, y=price)) +
    geom_point()
````

**Code für Python:**
```Python
ggplot(diamonds, aes(x='carat', y='price')) + \
    geom_point()
```

Über die Option *colour* der Funktion aes() kann nun auch noch eine weitere Variable angezeigt werden. Hierbei werden die verschiedenen Ausprägungen der Variable in unterschiedliche Farben angezeigt. In unserem Beispiel verwenden wir "color" als dritte Variable. 

````r
ggplot(diamonds, aes(x=carat, y=price, colour=color)) +
            geom_point()
````

** Code für Python: **
```Python
ggplot(diamonds, aes(x='carat', y='price', colour='color')) + \
            geom_point()
```

Das Verändern der Achsenbeschriftung kann mit `Scales` erreicht werden, als default ist dies immer die verwendete Variable. Ebenso kann in R der Name der Legende verändert werden, sowie in beiden Paketen ein Titel für die Grafik vergeben werden. 

> Variablenerklärung in labs():
> - **x**: Zugriff auf die Beschriftung der x-Achse
> - **y**: Zugriff auf die Beschriftung der y-Achse
> - **title**: Titel der Grafik
> - **colour (in R)**: Zugriff auf die Beschriftung der Legende

Wenden wir dies nun auf unser Beispiel an:

````r
ggplot(diamonds, aes(x=carat, y=price, colour=color)) +
            geom_point() + 
            labs(x="Karat", y="Preis", colour="Farbe", title="Diamantenliste")
````

** Code für Python:**
```Python
ggplot(diamonds, aes(x='carat', y='price', colour='color')) + \
            geom_point() + \ 
            labs(x="Karat", y="Preis", title="Diamantenliste")
```

Wollen wir nun die einzelnen Farben in einem eigenem Scatterplot betrachten, so kann dies über `facetting` gemacht werden. Hierbei wird die Funktion *facet_wrap* angefügt. In dieser Funktion gibt es auch wieder Unterschiede in der Schreibweise wie genau auf die Variable zugegriffen wird. Die Option *ncol* ist eine Zusatzoption und muss nicht zwingend angegeben werden, jedoch erhält man so eine größere Kontrolle über die Ausgabe der verschiedenen Grafiken.

````r
ggplot(diamonds, aes(x=carat, y=price)) +
    geom_point() +
    labs(x="Karat", y="Preis", title="Diamantenliste") +
    facet_wrap(~color, ncol=3)
````

**Code für Python:**
```Python
ggplot(diamonds, aes(x='carat', y='price')) + \
    geom_point() + \
    labs(x="Karat", y="Preis", title="Diamantenliste") + \
    facet_wrap('color', ncol=2)
```

Wie zu Erkennen, haben wir über `facetting` nun die Gruppierung einer dritten Variable in mehreren Grafiken dargestellt, anstatt dies in einer einzelnen Grafik einzuzeichnen. Das Einfärben der Verschiedenen Ausprägungen ist aber dennoch möglich, wenn wie vorhin in den `aesthetic mappings` die Option colour einfügt wird:

````r
ggplot(diamonds, aes(x=carat, y=price, colour=color)) +
    geom_point() +
    labs(x="Karat", y="Preis", colour="Farbe", title="Diamantenliste") +
    facet_wrap(~color, ncol=3)
````

**Code für Python:**
```Python
ggplot(diamonds, aes(x='carat', y='price', colour='color')) + \
    geom_point() + \
    labs(x="Karat", y="Preis", title="Diamantenliste") + \
    facet_wrap('color', ncol=2)
```

Wie bereits im theoretischen Teil erklärt wurde, können auch mehrere `Layer` in einem Plot vorkommen:

````r
ggplot(diamonds, aes(x=carat, y=price)) +
            geom_point() +
            labs(x="Karat", y="Preis", title="Diamantenliste") +
            geom_smooth()
````

**Code für Python:**
```Python
ggplot(diamonds, aes(x='carat', y='price')) + \
    geom_point() + \
    stat_smooth(color='red')
```

Natürlich sind Scatterplots nicht die einzigen Grafiken, die man mit den beiden Paketen erstellen kann,
hier zum Beispiel ein Histogramm:

````r
ggplot(diamonds, aes(x=price)) + 
    geom_histogram()
````

**Code für Python:**
```Python
ggplot(diamonds, aes(x='price')) + \
    geom_histogram()
```

Das Histogramm kann nun auch noch wie in der angezeigten Meldung verfeinert werden:

````r
ggplot(diamonds, aes(x=price,)) + 
    geom_histogram(binwidth=100)
````

**Code für Python:**
```Python
ggplot(diamonds, aes(x='price')) + \
    geom_histogram(binwidth=100)
```

Über die Option *fill* in den `aesthetic mappings` kann noch angegeben werden, wie viel von der Anzahl einer bestimmten Gruppe zuzuweisen ist.

````r
ggplot(diamonds, aes(x=price, fill=clarity)) + 
    geom_histogram(binwidth=100)
````

**Code für Python:**
```Python
ggplot(diamonds, aes(x='price', fill='clarity')) + \
    geom_histogram(binwidth=100)
```

Beide Pakete können also sehr umfangreich verwendet werden, um Grafiken in R und Python darzustellen. Hierbei kann manuell noch sehr viel umgestaltet werden. Die Optionen eine Grafik zu individualisieren sind so vielfältig, dass nicht alle in dieser Einführung der Pakete aufgelistet werden können. Jedoch ist ggplot2 von R so beliebt, dass schon mehrere Bücher erschienen sind, in denen Beschrieben wird, was genau alles mit ggplot2 erreicht werden kann.

Und zum Schluss noch die Spaßfunktion von Python:

```
